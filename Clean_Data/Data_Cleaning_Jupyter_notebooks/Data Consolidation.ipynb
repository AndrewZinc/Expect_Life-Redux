```python
# Import our dependencies
import pandas as pd
import numpy as np
from functools import reduce
```

```python
#  Import and read the input csv
g_ss_df = pd.read_csv('../../Clean_Data/Clean_CSV_Files/gov_systems-ss_systems-complete.csv', encoding='utf-8')
# Ensure the sss_depth column is small unsigned int type.
g_ss_df['sss_depth'] = g_ss_df['sss_depth'].astype(pd.Int8Dtype())
g_ss_df
```

```python
#  Import and read the input csv
md_df = pd.read_csv('../../Clean_Data/Clean_CSV_Files/MD_per_10k.csv', encoding='utf-8')
md_df = md_df.drop(['Country','Year'], axis=1)
md_df
```

```python
#  Import and read the input csv
cap_he_df = pd.read_csv('../../Clean_Data/Clean_CSV_Files/World_Bank_Capital_HE-1990_2022.csv', encoding='utf-8')
cap_he_df
```

```python
#  Import and read the input csv
wbdi_df = pd.read_csv('../../Clean_Data/Clean_CSV_Files/world_bank_dev_indicators-1990_2022.csv', encoding='utf-8')

wbdi_df['Total Labor Force (M)'] = wbdi_df['Total Labor Force'] / 1000000
wbdi_df['Total Population (M)'] = wbdi_df['Total Population'] / 1000000
wbdi_df = wbdi_df.drop(['Total Labor Force','Total Population'], axis=1)

wbdi_df
```

```python
#  Import and read the input csv
dcs_df = pd.read_csv('../../Clean_Data/Clean_CSV_Files/daily_calorie_supply.csv', encoding='utf-8')
dcs_df = dcs_df.drop(['Country','Year'], axis=1)
dcs_df
```

```python
# Merge the helper DataFrame with the original data to ensure the data range is complete.
frames = [g_ss_df, dcs_df, cap_he_df, md_df, wbdi_df]
complete_df = reduce(lambda left,right: pd.merge(left,right,how='left',on='Country_Year'),frames)

# Sort by 'Country' and 'Year'.
complete_df = complete_df.sort_values(['Country_Year'], axis=0)
```

```python
complete_df.columns
```

```python
# Rearrange the columns
complete_df = complete_df[['Country_Year','country','Year','government_type','sss_depth','system_type','Avg. Rainfall (mm/yr)','Population Density (people per sq. km)',
       'Total Population (M)','Total Labor Force (M)', 'LEx years','MD per 10k','Electricity Access (% Pop)','Current HE (% GDP)', "Current HE per capita (PPP Int'l $)", 'Capital HE (% GDP)',
       "Domestic general gov't HE (% GDP)","Domestic general gov't HE per capita (PPP Int'l $)","Domestic private HE per capita (PPP Int'l $)",
       'GDP growth per capita %',"GDP per capita (PPP Int'l $)",'Daily Caloric Supply','Tobacco use (% adults)','Alcohol Consumption per capita (liters)']]
```

```python
# Update the column names for brevity and clarity
complete_df = complete_df.rename(columns={'country':'Country','government_type':'Gov Type','sss_depth':'SSS Depth','system_type':'SSS Type',
            'Population Density (people per sq. km)':'Pop Density (#/sq km)','MD per 10k':'Doctors (#/10k pop)','Avg. Rainfall (mm/yr)':'Avg Rainfall (mm/yr)',
            "Current HE per capita (PPP Int'l $)":'Current HE per capita (PPP Intl $)',"Domestic general gov't HE (% GDP)":'Domestic General Gov HE (% GDP)',
            "Domestic general gov't HE per capita (PPP Int'l $)":'Domestic General Gov HE per capita (PPP Intl $)',"Domestic private HE per capita (PPP Int'l $)":'Domestic Private HE per capita (PPP Intl $)',
            "GDP per capita (PPP Int'l $)":'GDP per capita (PPP Intl $)','Alcohol Consumption per capita (liters)':'Alcohol Use per capita (liters)'})
```
