```python
# Import our dependencies
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
#  Import and read the input csv
dcs_df = pd.read_csv('../../Raw_Data/Daily_Caloric_Supply-vs-life-expectancy.csv')
dcs_df
```

```python
# Drop the extra columns
dcs_df = dcs_df.drop(['Code','Continent','Life expectancy - Sex: all - Age: at birth - Variant: estimates','Population (historical estimates)'], 1)
```

```python
dcs_df
```

```python
dcs_df = dcs_df[dcs_df['Year']>=1990]
dcs_df = dcs_df[dcs_df['Year']<=2023]
dcs_reduced_df = dcs_df
```

```python
# Open the master list of countries
countries_df = pd.read_csv('../../Clean_Data/master_country_list/country_profile_urls.csv')
countries_df = countries_df.drop(['profile_url'], 1)
```

```python
master_countries = countries_df['country'].tolist()
master_countries
```

```python
# Define a function to ensure any country names conform to the master list

def clean_countries(df):

    # Fuzzy matching: for each country in the input file, find the best match in the master list
    # If the match is good enough (e.g. score > 80), use the matched country name from the master list
    for idx, row in df.iterrows():
        best_match, score = process.extractOne(row['Entity'], master_countries)
        if score > 80:
            df.loc[idx,'Entity'] = best_match
        else:
            df.loc[idx,'Entity'] = None  # If no good match, set country to None
    # Return the cleaned and filtered data
    return df
```

```python
dcs_reduced_countries_df = clean_countries(dcs_reduced_df)
```

```python
dcs_reduced_countries_df
```

```python
# Create a list of Food Supply countries
ps_countries = dcs_reduced_countries_df['Entity'].tolist()
ps_countries
```

```python
# clean the FS data based on the SS country list
no_match = []

for country in ps_countries:
    if country in master_countries:
        continue
    else:
        no_match.append(country)
```

```python
# Likely need to do manual country name clean-up for accented characters and abbreviations.
no_match
```

```python
dcs_clean_df = dcs_reduced_countries_df.drop(dcs_reduced_countries_df[dcs_reduced_countries_df['Entity'].isin(no_match)].index.tolist())
dcs_clean_df
```

```python
dcs_clean_df.to_csv('../../Clean_Data/Clean_CSV_Files/daily_calorie_supply.csv', index=False)
```
